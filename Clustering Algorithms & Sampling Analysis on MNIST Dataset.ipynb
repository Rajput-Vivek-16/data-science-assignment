The given cost function of cluster is:

$$
\text{cost}(C) = \sum_{i=1}^{k} \frac{1}{|C_i|} \sum_{x, y \in C_i} \|x - y\|^2
$$
where:
- $|C_i|$ represents the number of points in cluster $C_i$.
- $C = \{C_1, C_2, \dots, C_k\}$ is a partitioning of the data points into $k$ clusters.

We know:
- The objective function of k-means is $
\sum_{i=1}^{k} \sum_{x \in C_i} \|x - c_i\|^2$
- The squared Educlidean distance between any two points x and y: $$
\|x - y\|^2 = \|x\|^2 + \|y\|^2 - 2 \langle x, y \rangle
$$
Taking sum the pairwise distances over all pairs (x,y) in $C_i$

$$
\sum_{x, y \in C_i} \|x - y\|^2 = \sum_{x, y \in C_i} (\|x\|^2 + \|y\|^2 - 2 \langle x, y \rangle)
$$

$$
\sum_{x, y \in C_i} \|x - y\|^2 = \sum_{x, y \in C_i} \|x\|^2 + \sum_{x, y \in C_i} \|y\|^2 - 2 \sum_{x, y \in C_i} \langle x, y \rangle
$$

$\sum_{x, y \in C_i} \|x\|^2 = |C_i| \sum_{x \in C_i} \|x\|^2$
and
$\sum_{x, y \in C_i} \|y\|^2 = |C_i| \sum_{x \in C_i} \|x\|^2$ because of symmetric in both

Put these in above expression

$$\sum_{x \in C_i} \|x - c_i\|^2 = \sum_{x \in C_i} \|x\|^2 - 2 |C_i| \langle c_i, c_i \rangle + |C_i| \|c_i\|^2$$

### **Comparison of Objective Functions in k-Means Clustering**  

Let $ X = \{x_1, x_2, \dots, x_n\} $ be the set of data points. Consider two cases for choosing cluster centers:  

- **Case 1:** Centers are restricted to be data points, denoted as $ C_{\text{data}} = \{c_1, c_2, \dots, c_k\} $. The objective function is:  
   $$
   F(C_{\text{data}}) = \sum_{i=1}^{n} \min_{1 \leq j \leq k} \|x_i - c_j\|^2
   $$  

- **Case 2:** Centers can be arbitrary points, denoted as $ C_{\text{arb}} = \{y_1, y_2, \dots, y_k\} $. The objective function is:  
   $$
   F(C_{\text{arb}}) = \sum_{i=1}^{n} \min_{1 \leq j \leq k} \|x_i - y_j\|^2
   $$  

Since arbitrary centers can always be chosen optimally, we have:  
$$
F(C_{\text{data}}) \leq F(C_{\text{arb}})
$$  


Using the **triangle inequality**, for any data point $ x_i $ and its closest centers $ y_j $ (from Case 2) and $ c_j $ (from Case 1), we obtain:  
$$
\| x_i - c_j \| \leq \| x_i - y_j \| + \| y_j - c_j \|
$$  
Squaring both sides:  
$$
\| x_i - c_j \|^2 \leq \| x_i - y_j \|^2 + 2 \| x_i - y_j \| \| y_j - c_j \| + \| y_j - c_j \|^2
$$  
Applying the identity $ 2ab \leq a^2 + b^2 $, we bound the cross term:  
$$
2 \| x_i - y_j \| \| y_j - c_j \| \leq \| x_i - y_j \|^2 + \| y_j - c_j \|^2
$$  
Substituting this into the inequality:  
$$
\| x_i - c_j \|^2 \leq 2 \| x_i - y_j \|^2 + 2 \| y_j - c_j \|^2
$$  
Summing over all data points:  
$$
F(C_{\text{data}}) = \sum_{i=1}^{n} \min_{1 \leq j \leq k} \| x_i - c_j \|^2
$$  
$$
\leq 2 \sum_{i=1}^{n} \min_{1 \leq j \leq k} \| x_i - y_j \|^2 + 2 \sum_{i=1}^{n} \min_{1 \leq j \leq k} \| y_j - c_j \|^2
$$  
Since the second term is bounded by $ kD^2 $ (where $ D $ is the maximum distance between any data point and an optimal center), we obtain:  
$$
F(C_{\text{data}}) \leq 2 F(C_{\text{arb}}) + 2k D^2
$$

For $F(C_{\text{arb}})$, we have:
$$
F(C_{\text{arb}}) = \sum_{i=1}^n \min_{1 \leq j \leq k} \| x_i - y_j|^2
$$

Let:
$$
D = \max_{i,j} \| x_i - y_j \|
$$

Then:
$$
\| x_i - y_j \| \leq D \quad \forall 1 \leq j \leq k
$$


$$
F(C_{\text{arb}}) \geq \sum_{i=1}^n D^2
$$


$$
F(C_{\text{arb}}) \geq n D^2
$$

$$
F(C_{\text{arb}}) \geq k D^2
$$

Finally:
$$
F(C_{\text{data}}) \leq 4 F(C_{\text{arb}})
$$

### **Conclusion**  
This result establishes that the cost of k-means clustering when restricting centers to data points is at most **4 times** the cost when choosing arbitrary centers:  
$$
{F(C_{\text{data}}) \leq 4 F(C_{\text{arb}})}
$$

Markov’s inequality states that for any non-negative random variable $X$ and any $t \geq 0$:

$$
P(X \ge t) \le \frac{\mathbb{E}[X]}{t}
$$

A random variable $X$ for which Markov’s inequality is **tight** that is an Bernoulli random variable.
$$
X = \begin{cases}
a, & \text{with probability } p \\
0, & \text{with probability } 1 - p
\end{cases}
$$
where $a > 0$.

We will show that for the above random variable, markov's inequality is tight.

The expectation of $X$ is:
$$E[X] = a \cdot p + 0 \cdot (1 - p) = a p$$

The probability $P(X \geq t)$ is:
$$P(X \geq t) = P(X = t) = p$$

Now, apply Markov's inequality gives:
- Case 1: $t=a$
$$
P(X \ge t) \le \frac{\mathbb{E}[X]}{t}
$$

$$
P(X \ge t) \le \frac{ap}{t} = p
$$

$$
P(X \ge t) = \frac{\mathbb{E}[X]}{t}
$$
Hence, markov is tight in this case.
- Case 2: $t \in (0, a)$
$$
P(X \ge t) = \frac{\mathbb{E}[X]}{t}
$$
In this case also, markov is tight because of bernoulli distribution.

Hence, the bernoulli distribution is an example where markov inequality is tight.

---

### **Why do we use Other Inequalities like Chebyshev and Chernoff?**
Even though Markov’s inequality is useful, it is often **loose** because we often use complex distribution rather than simpler in practical use and markov depends on the expectation of $X$. Therefore, we use other inequalities  which does not depend on expectation of $X$ and tigher than markov.

A tigher bound can be obtained using variance or exponential functions:
- **Chebyshev’s Inequality**:
   - Provides a tighter bound when variance is low.
   - Useful when dealing with deviations around the mean.

- **Chernoff Bounds**:
   - Exponential bounds on tail probabilities, much stronger than Markov’s inequality.
   - Works well for sums of independent random variables.

Thus, while Markov’s inequality is useful in general, it is often not the **best** choice than other inqualities.

```python
from keras.datasets import mnist
from sklearn.metrics import rand_score
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(f'Datasets contains {x_test.shape[0]} test images, each image has a height of {x_test.shape[1]} and width of {x_test.shape[1]} pixels.')
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f"Label: {y_test[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## a) K-Means with kmeans++

```python
x_test = x_test.reshape(len(x_test), -1)

def mykmeanspp(x_test, n_centroids):
    n_images = x_test.shape[0]
    n_pixels = x_test.shape[1]
    centers = np.zeros((n_centroids, n_pixels))
    centers[0] = x_test[np.random.randint(0, n_images)]

    for i in range(1, n_centroids):
        sq_dist_each_center = np.array([np.sum((centers[k] - x_test) ** 2, axis=1) for k in range(i)])
        sq_dist = np.min(sq_dist_each_center, axis=0)
        prob_dist = sq_dist / np.sum(sq_dist)
        max_prob_index = np.random.choice(n_images, p=prob_dist)
        centers[i] = x_test[max_prob_index]

    return centers

centers = mykmeanspp(x_test, 10)

def kmean(centers, x_test):
    n_images = x_test.shape[0]
    dist_each_center = np.array([np.linalg.norm(x_test - center, axis=1) for center in centers])
    ind_cluster = np.argmin(dist_each_center, axis=0)
    clusters = [[] for _ in range(len(centers))]
    for i in range(n_images):
        clusters[ind_cluster[i]].append(x_test[i])

    return clusters, ind_cluster

clusters, ind_cluster = kmean(centers, x_test)
rand_index = rand_score(y_test, ind_cluster)
print(f"Rand Index: {rand_index}")
```

## b) k-Center Clustering

```python
def kcenter(x_test, n_centroids):
    n_images = x_test.shape[0]
    n_pixels = x_test.shape[1]

    centers = np.zeros((n_centroids, n_pixels))
    first_index = np.random.randint(0, n_images)
    centers[0] = x_test[first_index]

    for i in range(1, n_centroids):
        distances = np.min([np.linalg.norm(x_test - centers[j], axis=1) for j in range(i)], axis=0)
        next_center = np.argmax(distances)
        centers[i] = x_test[next_center]

    dist_each_center = np.array([np.linalg.norm(x_test - centers[k], axis=1) for k in range(n_centroids)])
    ind_cluster = np.argmin(dist_each_center, axis=0)

    clusters = [[] for _ in range(n_centroids)]
    for i in range(n_images):
        clusters[ind_cluster[i]].append(x_test[i])
    return clusters, ind_cluster

clusters, ind_cluster = kcenter(x_test, 10)

rand_index = rand_score(y_test, ind_cluster)
print(f"Rand-index of the k-center clustering: {rand_index}")
```

## C)  Single Linkage Agglomeration

```python
agglo_cluster = AgglomerativeClustering(n_clusters=10, linkage='single')
predicted_labels = agglo_cluster.fit_predict(x_test[:100])

rand_index = rand_score(y_test[:100], predicted_labels)
print(f"Rand-index of the Single Linkage Agglomeration clustering: {rand_index}")
```

## D) Rank-k approximation of the training data matrix

```python
x_train = x_train.reshape(len(x_train), -1)[:100]
y_train = y_train[:100]

def rank_k_approximation(k, data):
    u, sigma, v = np.linalg.svd(data, full_matrices=False)
    u_k = u[:, :k]
    sigma_k = np.diag(sigma[:k])
    x_approx = np.dot(u_k, sigma_k @ v[:k, :])
    return x_approx

rank_values = [2, 5, 10]
rand_ind_kmeans = []
rand_ind_kcenter = []

for k in rank_values:
    x_approx = rank_k_approximation(k, x_train)
    centers = mykmeanspp(x_approx, 10)
    clusters, ind_cluster = kmean(centers, x_approx)
    rand_ind_kmeans.append(rand_score(y_train, ind_cluster))

    x_approx = rank_k_approximation(k, x_train)
    clusters, ind_cluster_kcenter = kcenter(x_approx, 10)
    rand_ind_kcenter.append(rand_score(y_train, ind_cluster_kcenter))

for i, rank_k in enumerate(rank_values):
    print(f"Rank-k = {rank_k}:")
    print(f"  K-Means Rand Index: {rand_ind_kmeans[i]:.4f}")
    print(f"  K-Center Rand Index: {rand_ind_kcenter[i]:.4f}")
```

**Algorithm:**
- Define the Population and Sample
  - Let the total population be N = $10^6$
  - First, we take random samples of size n from the population.
  - Now out of n, we count the number of coffee drinkers, let say $S$.
  - The probability of coffee drinkers that is $\hat{p} = \frac{S}{n}$.


- Sample Random Individuals:

$X_i$ is an random variable for the $i^{th}$ person in the sample:
$$
X_i =
\begin{cases}
1, & \text{if $i^{th}$ person drinks coffee} \\
0, & \text{otherwise}
\end{cases}
$$

The total number of coffee drinkers in the sample is
$$
S = X_1 + X_2 + \cdots + X_n.
$$

The sample proportion of coffee drinker is
 $$
   \hat{p} = \frac{1}{n} \sum_{i=1}^{n} X_i = \frac{S}{n}
   $$

- Error Bounds using Chernoff Bound:
The expected value of is $E[\hat{p}] = p$

  - Upper Tail:  (For small values of delta)
      $$
    P\Bigl(S \ge (1+\delta)np\Bigr) \le e^{-\frac{\delta^2np}{2}}
      $$

  - Lower Tail
      $$
    P\Bigl(S \le (1-\delta)np\Bigr) \le e^{-\frac{\delta^2np}{2}}
      $$


  From above two bound conditions we obtained:
$P\Bigl(|S - np| \ge \delta np\Bigr) \le 2 e^{-\frac{\delta^2np}{2}}$

  Divide the LHS by np in the probability we get :

  $$
    P\Bigl(\frac{|\frac{S}{n} - p|}{p} \ge \delta \Bigr) \le 2 e^{-\frac{\delta^2np}{2}}
      $$

  Here $\frac{|\frac{S}{n} - p|}{p}$ is the relative error in the computed probabilty using sampling.
  So we set delta accordingly

  For any δ > 0, the probability that $ \hat{p} $ deviates from $ p $ satisfies:  

   $$
    P(|\hat{p} - p| \geq \delta p) \leq 2 e^{-\frac{\delta^2np}{2}}
   $$  

  - Choosing Sample Size:

    To get an error bound $δ$ with probability at least 99%, we set:
      
$$2 \exp\left( -\frac{\delta^2 np}{2} \right) \leq 0.01$$  

After simplifying, we get the expression of $n$:  

$$n \geq \frac{2 \ln(200)}{\delta^2 p}$$  

For **10% relative error** ($\delta = 0.1$):  

$$n \geq \frac{2 \ln(200)}{(0.1)^2 \times 0.01} = 106000$$  
$$n \geq 106000$$  

For **5% relative error** ($\delta = 0.05$):  

$$n \geq \frac{2 \ln(200)}{(0.05)^2 \times 0.01} = 424000$$

$$n \geq 424000$$  

Therefore,  
- Sample of 106,000 people, we can get the fraction with a 10% relative error with 99% probability.  
- Sample of 424,000 people, we can get the fraction with a a 5% relative error with 99% probability.
